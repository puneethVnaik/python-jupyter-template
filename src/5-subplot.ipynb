### Exercise 5: Subplot Layout

```python
# TODO: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.


import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]


fig, axs = plt.subplots(2, 2, figsize=(10, 10))


axs[0, 0].plot(x, y, marker='o', color='blue', label='y = 2x')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 0].grid(True)
axs[0, 0].legend()


axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'purple'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')


axs[1, 0].scatter(x, y, color='orange', label='Data Points')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')
axs[1, 0].legend()


axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'coral'])
axs[1, 1].set_title('Pie Chart')


plt.tight_layout()


plt.show()
```
